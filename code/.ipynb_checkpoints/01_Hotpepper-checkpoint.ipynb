```python
import requests
```

```python
from api_key import hotpepper_api_key
```

```python
URL = 'http://webservice.recruit.co.jp/hotpepper/gourmet/v1/'
```

```python
params = {
    'key': hotpepper_api_key.API_KEY,
    'keyword': '沖縄',
    'format': 'json',
    'count': 100
}
```

```python
res = requests.get(URL, params)
```

```python
result = res.json()
```

```python
items = result['results']['shop']
```
